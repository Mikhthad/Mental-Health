The Impact of Remote Work on Mental Health dataset provides a detailed view of the relationship between remote working conditions and the mental well-being of employees from various industries and regions. The dataset captures multiple aspects of work-life dynamics, including job roles, work location, mental health conditions, productivity changes, and social isolation, among others. Each employee is uniquely identified and categorized based on demographic factors such as age, gender, and years of experience, allowing for granular insights into how different groups of employees respond to remote work.


It also includes factors such as the number of virtual meetings attended, work-life balance ratings, stress levels, access to mental health resources, physical activity, and sleep quality. These factors combine to provide a holistic view of how remote work impacts mental health and job satisfaction.

## Column Description


Employee_ID: Unique identifier for each employee (object).

Age: Age of the employee (int64).

Gender: Gender of the employee (object).

Job_Role: Job position or role of the employee (object).

Industry: Industry in which the employee works (object).

Years_of_Experience: Total years of work experience (int64).

Work_Location: Type of work location (e.g., remote, in-office, hybrid) (object).

Hours_Worked_Per_Week: Number of hours worked per week (int64).

Number_of_Virtual_Meetings: Number of virtual meetings attended per week (int64).

Work_Life_Balance_Rating: Rating of work-life balance (int64).

Stress_Level: Stress level of the employee (object).

Mental_Health_Condition: Employee's mental health condition (object).

Access_to_Mental_Health_Resources: Whether the employee has access to mental health resources (object).

Productivity_Change: Change in productivity due to remote work (object).

Social_Isolation_Rating: Rating of how isolated the employee feels (int64).

Satisfaction_with_Remote_Work: Satisfaction level with remote work (object).

Company_Support_for_Remote_Work: Rating of the company's support for remote work (int64).

Physical_Activity: Frequency or type of physical activity (object).

Sleep_Quality: Rating of sleep quality (object).

Region: Region where the employee is located (object).

## Job Role Distribution

```python
plt.figure(figsize = (12,6))
sns.countplot(df['Industry'],palette = 'viridis')
plt.title('Job Role Distribution')
plt.xlabel('Frequency')
plt.show()
```

IT and Finance Dominance: These sectors have the highest number of employees, indicating a significant focus on these fields within the dataset.

```python
plt.figure(figsize = (12,6))
sns.countplot(x = 'Work_Location',hue = 'Productivity_Change',data = df)
plt.title('Work Location vs. Productivity Change')
plt.ylabel('Frequency')
plt.show()
```

* Hybrid Work: The majority of employees in a hybrid work setting report an decrease in productivity. However, there are also notable numbers reporting a no change and decrease.
* Remote Work: Most employees working remotely also experience an decrease in productivity ,Then the Increase and No Change is almost same.
* Onsite Work: Productivity changes are less pronounced for onsite workers,The most reported of these was no change .Then decrease then increase 

```python
plt.figure(figsize = (12,6))
sns.countplot(x = 'Work_Location',hue = 'Stress_Level',data = df)
plt.title("Stress Level Distribution Across Different Work Locations")
plt.show()
```

The distribution of stress levels (Low, Medium, and High) is relatively consistent across Hybrid, Remote, and Onsite work locations, indicating that work location alone may not be a significant factor in influencing employee stress. Instead, stress levels seem evenly distributed, with "Medium" stress slightly more prevalent in Remote work, while "High" stress remains consistently present across all locations.

```python
plt.figure(figsize = (12,6))
sns.countplot(x = 'Mental_Health_Condition',hue = 'Work_Location',data = df)
plt.title('Distribution of Mental Health Conditions by Work Location')
plt.ylabel('Frequency')
plt.show()
```

Remote work is associated with higher levels of anxiety, while depression and burnout are consistent across all work setups. This calls for targeted mental health support for remote employees and a comprehensive approach to addressing mental well-being in all work environments.

```python
plt.figure(figsize = (12,6))
sns.countplot(x = 'Mental_Health_Condition',hue = 'Productivity_Change',data = df)
plt.title("Influence of Mental Health Condition on Productivity Change")
plt.show()
```

Depression and burnout are the most likely to reduce productivity, while anxiety often results in no immediate change. These insights underscore the need for tailored mental health support, with specific attention to depression and burnout management to sustain long-term employee productivity.
